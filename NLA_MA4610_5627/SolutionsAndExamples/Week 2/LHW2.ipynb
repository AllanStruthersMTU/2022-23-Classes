One way to print pretty latex $\int_0^\pi$.

```julia
using LinearAlgebra
(m,n)=(400,100) #40,000 gives 0.1 seconds. 
A = rand(m,n)
@time (Q, R) = qr(A)
(size(Q),size(Matrix(Q)))
```

Something is funny about Q!

```julia
QTS = Matrix(Q)
norm(QTS'*QTS - I)
```
